```python
import pandas as pd
data_submits=pd.read_csv('https://stepik.org/media/attachments/course/4852/submissions_data_train.zip')
```

```python
data_submits.sort_values(by='step_id')
```

```python
data1=data_submits.iloc[data_submits['timestamp'].idxmax()]
```

```python
data_max_timestamp=data_submits.groupby('user_id')['timestamp'].max()
max_tmstmp=pd.DataFrame(data_max_timestamp)
```

```python
max_tmstmp=max_tmstmp.rename(columns={'timestamp':'max_timestamp'})
```

```python
data_submits_with_max=data_submits.merge(max_tmstmp, how='inner', on='user_id')
```

```python
wrong_last_answr=data_submits_with_max.query("(submission_status=='wrong') & (timestamp == max_timestamp)")
```
